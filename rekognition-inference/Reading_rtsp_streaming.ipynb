```python
import boto3
import cv2
import matplotlib.pyplot as plt
```

```python
rtsp_addr = "rtsp://lb1.sstlive.com:1935/live/HIKtest"
cap = cv2.VideoCapture(rtsp_addr)
```

```python
frame_id=0
while(True):
  ret, frame = cap.read()

  if ret == True: 

    # Display the resulting frame    
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.title('Frames')
    plt.show()
    frame_id+=1
    # Press Q on keyboard to stop
    if cv2.waitKey(1) & frame_id==5:
      break

  # Break the loop
  else:
    break  
 
```
